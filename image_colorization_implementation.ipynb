```python
import tensorflow as tf
from PIL import Image
import numpy as np
import os
from tensorflow import keras
import matplotlib.pyplot as plt
from skimage.color import rgb2lab , lab2rgb
from skimage.io import imread , imshow
from tensorflow.keras.utils import img_to_array, load_img
from tensorflow.python.ops.numpy_ops import np_config
np_config.enable_numpy_behavior()
img1_color = []
path = "/content/images.jpg"
img = img_to_array(
    load_img(path, target_size = (256,256,3)))/255
img1_color.append(img)
img1_color = rgb2lab(img1_color)[:,:,:,0]
img1_color = img1_color.reshape(img1_color.shape+(1,))

def load_model():
  model_file_path = '/content/my_model.h5'
  model = keras.models.load_model("/content/mymodel2.h5")
  return model
model = load_model()

result = model.predict(img1_color)
output1 = model.predict(img1_color)
output1 = output1*128
result = np.zeros((256,256,3))
result[:,:,0] = img1_color[0][:,:,0]
result[:,:,1:] = output1[0]
imshow(lab2rgb(result))
```
